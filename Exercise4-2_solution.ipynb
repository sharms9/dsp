```python
#Import required libararies...
import random
import thinkstats2
import thinkplot
import matplotlib.pyplot as plt
```

```python
#Generate a list of 1000 random numbers...
randomnum = [random.random() for _ in range(1000)]
```

```python
#Spot check what the list looks like (first 30 entries, last 30 entries)...
print randomnum[:30]
print
print randomnum[-30:]
```

```python
#Let's look at the pmf next...
#First generate the pmf values...
values_pmf = thinkstats2.Pmf(randomnum)
#Then generate the pmf plot...
%matplotlib inline
plt.figure(figsize=(20,10))
thinkplot.Pmf(values_pmf, linewidth = 0.1)
plt.title('PMF Values')
```

```python
#Now let's look at the cdf...
```

```python
#Start by generating cdf values
values_cdf = thinkstats2.Cdf(randomnum)
#Then generate the cdf plot like we did with pmf...
plt.figure(figsize=(20,10))
thinkplot.Cdf(values_cdf)
plt.title('CDF Values')
```

The distribution is uniform, slope = 1.
